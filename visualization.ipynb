# 이상치 확인

```python
list_boxplots = ['Year_Birth', 'Income', 'Recency','MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

fig, ax = plt.subplots(ncols=4, nrows=4, figsize=(16, 15))

cont = 0
for i in range(4):
    for j in range(4):
        if cont < len(list_boxplots):
            sns.boxplot(df[list_boxplots[cont]], ax=ax[i][j], color='#4CAF50')
            ax[i][j].set_title(list_boxplots[cont])
            cont = cont + 1
        else:
            ax[i][j].axis('off')

plt.tight_layout()
plt.show()
```

# 범주형 & 연속형 변수

```python
# 그림 크기 설정
plt.figure(figsize=(20, 35))

# subplot에 대한 변수와 그래프 정의
variables = [
    ('Education', 'Variable Education'),
    ('Marital_Status', 'Variable Marital_Status'),
    ('Kidhome', 'Variable Kidhome'),
    ('Teenhome', 'Variable Teenhome'),
    ('NumStorePurchases', 'Variable NumStorePurchases'),
    ('NumDealsPurchases', 'Variable NumDealsPurchases'),
    ('NumWebPurchases', 'Variable NumWebPurchases'),
    ('NumCatalogPurchases', 'Variable NumCatalogPurchases'),
    ('NumWebVisitsMonth', 'Variable NumWebVisitsMonth'),
    ('Response', 'Variable Response'),
    ('Complain', 'Variable Complain')
]

# subplot에 각 변수의 카운트 플롯 그리기
for i, (variable, title) in enumerate(variables, start=1):
    plt.subplot(6, 2, i)
    plt.gca().set_title(title)
    sns.countplot(x=variable, palette='Set2', data=df)

# subplot 간 간격 조정
plt.tight_layout()
plt.show()
```

```python
# 그림 크기 설정
plt.figure(figsize=(20, 35))

# subplot에 대한 변수와 그래프 정의
variables = [
    ('Education', 'Variable Education'),
    ('Marital_Status', 'Variable Marital_Status'),
    ('Kidhome', 'Variable Kidhome'),
    ('Teenhome', 'Variable Teenhome'),
    ('NumStorePurchases', 'Variable NumStorePurchases'),
    ('NumDealsPurchases', 'Variable NumDealsPurchases'),
    ('NumWebPurchases', 'Variable NumWebPurchases'),
    ('NumCatalogPurchases', 'Variable NumCatalogPurchases'),
    ('NumWebVisitsMonth', 'Variable NumWebVisitsMonth'),
    ('Response', 'Variable Response'),
    ('Complain', 'Variable Complain')
]

# subplot에 각 변수의 카운트 플롯 그리기
for i, (variable, title) in enumerate(variables, start=1):
    plt.subplot(6, 2, i)
    plt.gca().set_title(title)
    sns.countplot(x=variable, hue='Response', data=df, palette='Set2', dodge=True)

# subplot 간 간격 조정
plt.tight_layout()
plt.show()
```
